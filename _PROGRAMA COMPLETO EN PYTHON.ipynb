```python
def tour_extrangeros():
    # Variables para seguimiento
    cantidad_hotel_p = cantidad_hotel_E = cantidad_hotel_C = 0
    pasajeros_b = pasajeros_v = pasajeros_oltursa = 0
    importe_bus_cama_O = importe_bus_cama_C = importe_bus_cama_L = 0
    importe_vip_O = importe_vip_C = importe_vip_L = 0
    suma_hotel_p = suma_hotel_E = suma_hotel_C = 0
    transporte = ' '

    while transporte != 'F':
        # Selección del tipo de transporte
        while True:
            transporte = input("Ingrese el tipo de transporte (B: bus cama; V: servicio VIP; F: salir): ").upper()
            if transporte in ['B', 'V', 'F']:
                break
            else:
                print("Ingrese un tipo de transporte válido.")

        if transporte == 'F':
            break

        # Selección de la empresa de transporte
        while True:
            empresa = input("Ingrese la empresa en la que viaja (O: Oltursa; C: Cruz del Sur; L: Línea): ").upper()
            if empresa in ['O', 'C', 'L']:
                break
            else:
                print("Datos de empresa incorrectos. Intente nuevamente.")

        # Selección del hotel
        while True:
            hotel = input("Ingrese el hotel en el que se hospeda (P: Los Pinos; E: Emperador; C: Cañaveral): ").upper()
            if hotel in ['P', 'E', 'C']:
                break
            else:
                print("Datos de hotel incorrectos. Intente nuevamente.")

        # Ingreso del número de días del viaje
        while True:
            try:
                dias_viaje = int(input("Número de días del viaje: "))
                if dias_viaje >= 3:
                    break
                else:
                    print("El número de días debe ser al menos 3.")
            except ValueError:
                print("Por favor, ingrese un número entero válido.")

        # Procesamiento del transporte
        if transporte == 'B':
            pasajeros_b += 1
            if empresa == 'O':
                pasajeros_oltursa += 1
                importe_bus_cama_O += 150
            elif empresa == 'C':
                importe_bus_cama_C += 200
            elif empresa == 'L':
                importe_bus_cama_L += 100
        elif transporte == 'V':
            pasajeros_v += 1
            if empresa == 'O':
                pasajeros_oltursa += 1
                importe_vip_O += 230
            elif empresa == 'C':
                importe_vip_C += 300
            elif empresa == 'L':
                importe_vip_L += 0

        # Procesamiento del hotel
        if hotel == 'P':
            cantidad_hotel_p += 1
            suma_hotel_p += (20 * 3.90)* dias_viaje
        elif hotel == 'E':
            cantidad_hotel_E += 1
            suma_hotel_E += (30 * 3.90)* dias_viaje
        elif hotel == 'C':
            cantidad_hotel_C += 1
            suma_hotel_C += (30 * 3.90) * dias_viaje

    # Resultados finales
    total_pasajeros = pasajeros_b + pasajeros_v
    porcentaje_oltursa = (pasajeros_oltursa / total_pasajeros) * 100 if total_pasajeros > 0 else 0

    # Determinar el hotel con más ingresos
    max_ingresos = max(suma_hotel_p, suma_hotel_E, suma_hotel_C)
    hoteles_mayores = []
    if suma_hotel_p == max_ingresos:
        hoteles_mayores.append("Los Pinos")
    if suma_hotel_E == max_ingresos:
        hoteles_mayores.append("Emperador")
    if suma_hotel_C == max_ingresos:
        hoteles_mayores.append("Cañaveral")

    # Impresión de resultados
    print("\nRESULTADOS")
    print(f"Cantidad de viajeros que se alojan en hoteles de USD 30: {cantidad_hotel_E + cantidad_hotel_C}")
    print(f"Porcentaje de pasajeros que eligieron Oltursa: {porcentaje_oltursa:.2f}%")

    print("\nReporte de ingresos de cada empresa:")
    print(f"{'*' * 30}")
    print(f"  OLTURSA:")
    print(f"    Bus Cama     : {importe_bus_cama_O:.2f} USD")
    print(f"    Servicio VIP : {importe_vip_O:.2f} USD")
    print(f"  CRUZ DEL SUR:")
    print(f"    Bus Cama     : {importe_bus_cama_C:.2f} USD")
    print(f"    Servicio VIP : {importe_vip_C:.2f} USD")
    print(f"  LÍNEA:")
    print(f"    Bus Cama     : {importe_bus_cama_L:.2f} USD")
    print(f"    Servicio VIP : {importe_vip_L:.2f} USD")

    print(f"El hotel con más ingresos fue: {', '.join(hoteles_mayores)}")
    print(f"El monto total fue de: {max_ingresos:.2f} USD\n")

# Ejecutar el programa
tour_extrangeros()
```
