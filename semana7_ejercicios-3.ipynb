```python
def Tarjetas():
    jovenes_mujeres_viudas = 0
    hombres_solteros_divorciados = 0
    cantidad_hombres = 0
    cantidad_mujeres = 0
    tarjetas = int (input ("ingrese numero de tarjetas : "))
    for i in range (1, tarjetas +1):
        print(f"tajeta {i}")
        while True:
            try:
                sexo = input("ingrese sexo ( F , M): ").upper()
                if sexo in ['F', 'M']:
                    break
                else:
                    print("sexo incorreto")
            except ValueError:
                print("datos incorrectos...")
        while True:
            try:
                edad = int (input("Edad: "))
                if 1<= edad <= 100:
                    break
                else:
                    print("Edad fuera de rango ( 1- 100 )")
            except ValueError:
                print("datos incorrectos..")        
        while True:
            try:
                estado_civil = input("Estado civil (A : soltero B: casado C:Vviudo D: divorciado) : ").upper()
                if estado_civil in ['A', 'B', 'C', 'D']:
                    break
                else:
                    print("Estado civil incorrectos..")
            except ValueError :
                print("datos incorrectos..")
        if sexo =='F' and estado_civil == 'C' and  16<= edad<=21:
            jovenes_mujeres_viudas +=1

        if sexo == 'F':
            cantidad_mujeres +=1
        else:
            cantidad_hombres+=1

        if sexo == 'M' and estado_civil in ['A', 'D']:
            hombres_solteros_divorciados+=1

    porcentaje_hombres = (cantidad_hombres/tarjetas)*100
    porcentaje_mujeres = (cantidad_mujeres/ tarjetas)*100
      

    #resultados 
    print("RESULTADOS")
    print(f"jovenes viudas entre 16 y 21 años : {jovenes_mujeres_viudas}")
    print(f"porcentaje de hombre   : {porcentaje_hombres:.2f}%")
    print(f"porcentaje de mujeres  : {porcentaje_mujeres:.2f}%")
    print(f"hombres mayores de 30 años solteros y divorciados : {hombres_solteros_divorciados}")
    #ejecutar funcion
Tarjetas()
```

```python
def encuesta():
    suma_niños = suma_jovenes = suma_adultos = suma_adulto_mayor = 0
    niños = jovenes = adultos = adulto_mayor = 0

    personas = int(input("Ingrese el número de personas a registrar: "))
    for i in range(1, personas + 1):
        print(f"\nPersona {i}")
        
        # Validar peso
        while True:
            try:
                peso = float(input("Ingrese el peso: "))
                if peso > 0:
                    break
                else:
                    print("Peso incorrecto. Debe ser mayor a 0.")
            except ValueError:
                print("Dato incorrecto. Ingrese un número válido.")

        # Validar edad
        while True:
            try:
                edad = int(input("Ingrese su edad: "))
                if edad > 0:
                    break
                else:
                    print("Edad incorrecta. Debe ser mayor a 0.")
            except ValueError:
                print("Dato incorrecto. Ingrese un número válido.")

        # Clasificar y acumular datos
        if 1 <= edad <= 12:
            suma_niños += peso
            niños += 1
        elif edad <= 29:
            suma_jovenes += peso
            jovenes += 1
        elif edad <= 59:
            suma_adultos += peso
            adultos += 1
        else:
            suma_adulto_mayor += peso
            adulto_mayor += 1

    # Calcular y mostrar promedios
    print("\nResultados:")
    if niños > 0:
        promedio_niños = suma_niños / niños
        print(f"Promedio de peso niños          : {promedio_niños:.2f} kg")
    else:
        print("No se registraron niños.")

    if jovenes > 0:
        promedio_jovenes = suma_jovenes / jovenes
        print(f"Promedio de peso jóvenes        : {promedio_jovenes:.2f} kg")
    else:
        print("No se registraron jóvenes.")

    if adultos > 0:
        promedio_adultos = suma_adultos / adultos
        print(f"Promedio de peso adultos        : {promedio_adultos:.2f} kg")
    else:
        print("No se registraron adultos.")

    if adulto_mayor > 0:
        promedio_adulto_mayor = suma_adulto_mayor / adulto_mayor
        print(f"Promedio de peso adultos mayores: {promedio_adulto_mayor:.2f} kg")
    else:
        print("No se registraron adultos mayores.")

# Ejecutar la función
encuesta()
```

```python
def sueldo_vendedores():
    suma_comisiones = 0
    personas = int(input("Ingrese el número de vendedores: "))
    
    # Itera sobre cada vendedor
    for i in range(1, personas + 1):
        print(f"Vendedor {i}")
        print("*" * 18)
        
        nombre = input("Ingrese nombre del vendedor: ").title()
        sueldo_base = float(input(f"Ingrese el sueldo base de {nombre}: "))
        suma_comisiones = 0  # Resetea la suma de comisiones por cada vendedor
        ventas = int(input(f"Ingrese el número de ventas realizadas por {nombre} (3): "))
        
        for j in range(1, ventas + 1):
            venta = float(input(f"Venta {j}: "))
            comision = venta * 0.10  # Comisión del 10%
            suma_comisiones += comision  # Acumula las comisiones
        
        # Calcula el sueldo neto
        sueldo_neto = sueldo_base + suma_comisiones
        
        # Muestra el resumen del sueldo
        print(f"Resumen del sueldo de {nombre}:")
        print(f"Comisiones por venta: {suma_comisiones}")
        print(f"Sueldo neto: {sueldo_neto}")
        print("-" * 30)

# Ejecutar la función
sueldo_vendedores()
```

```python
# Función para validar el número de alumnos
def validar_numero_de_alumnos():
    while True:
        N = int(input("Ingrese el número de alumnos (debe ser entre 1 y 40): "))
        if 1 <= N <= 40:
            return N
        else:
            print("Número inválido. El número de alumnos debe ser entre 1 y 40.")

# Función principal
def main():
    N = validar_numero_de_alumnos()
    total_alumnos = 40
    bien_tres_preguntas = 0
    aprobados = 0
    mal_pregunta_3 = 0
    no_rindieron = 0

    for i in range(N):
        respuestas_correctas = 0
        print(f"\nAlumno {i + 1}:")

        # Ingresar respuestas de las 3 preguntas
        respuesta_1 = int(input("Pregunta 1 (1-5): "))
        respuesta_2 = int(input("Pregunta 2 (1-5): "))
        respuesta_3 = int(input("Pregunta 3 (1-5): "))

        # Verificar las respuestas correctas
        if respuesta_1 == 3:
            respuestas_correctas += 1
        if respuesta_2 == 1:
            respuestas_correctas += 1
        if respuesta_3 == 4:
            respuestas_correctas += 1

        # Verificar si el alumno hizo bien las tres preguntas
        if respuestas_correctas == 3:
            bien_tres_preguntas += 1

        # Verificar si el alumno aprobó (al menos 2 respuestas correctas, una de ellas la 2)
        if respuestas_correctas >= 2 and respuesta_2 == 1:
            aprobados += 1

        # Verificar si el alumno hizo mal la pregunta 3
        if respuesta_3 != 4:
            mal_pregunta_3 += 1

    # Calcular cantidad de alumnos que no rindieron el examen
    no_rindieron = total_alumnos - N

    # Calcular el porcentaje de alumnos que hizo bien las tres preguntas
    porcentaje_bien_tres_preguntas = (bien_tres_preguntas / N) * 100

    # Imprimir los resultados
    print("\nResultados:")
    print(f"Porcentaje de alumnos que hizo bien las tres preguntas: {porcentaje_bien_tres_preguntas:.2f}%")
    print(f"Cantidad de alumnos aprobados: {aprobados}")
    print(f"Cantidad de alumnos que hizo mal la pregunta 3: {mal_pregunta_3}")
    print(f"Cantidad de alumnos que no rindieron el examen: {no_rindieron}")

# Ejecutar el programa
main()

```
