```python
from random import randint

def generar_mostrar():
    while True:
        try:
            alumnos = int(input("Ingrese el número de alumnos (1-30): "))
            if 1 <= alumnos <= 30:
                break
            else:
                print("Dato fuera de rango")
        except ValueError:
            print("Dato incorrecto, ingrese un número válido")
    puntaje_alumnos = []
    tema_alumnos = []
    for i in range(alumnos):
        Tema = randint(1, 3)
        puntaje = randint(0, 50)
        puntaje_alumnos.append(puntaje)
        tema_alumnos.append(Tema)
    return puntaje_alumnos, tema_alumnos, alumnos

def mostrar_datos(puntaje_alumnos, tema_alumnos, alumnos):
    print("\nDatos de los alumnos en forma vertical:")
    print("Alumno".center(10), "Tema".center(10), "Puntaje".center(10))
    print("-" * 30)
    for i in range(alumnos):
        print(f"{' '*3} {i+1:<10} {tema_alumnos[i]:<10} {puntaje_alumnos[i]:<10}")
    print("-" * 30)

def promedio_puntaje(puntaje_alumnos, tema_alumnos):
    promedio = sum(puntaje_alumnos) / len(puntaje_alumnos)
    print(f"\nPromedio de puntajes: {promedio:.2f}")
    print("Alumnos con puntaje mayor al promedio:")
    for i, puntaje in enumerate(puntaje_alumnos):
        if puntaje > promedio:
            print(f"  Alumno {i+1} -> Tema: {tema_alumnos[i]} Puntos: {puntaje}")

def mayor_puntaje(puntaje_alumnos, tema_alumnos):
    maximo = max(puntaje_alumnos)
    print(f"\nMayor puntaje: {maximo}")
    for i, puntaje in enumerate(puntaje_alumnos):
        if puntaje == maximo:
            print(f"  Alumno {i+1} -> Tema: {tema_alumnos[i]} Puntos: {puntaje}")

def main():
    puntaje_alumnos, tema_alumnos, alumnos = generar_mostrar()
    mostrar_datos(puntaje_alumnos, tema_alumnos, alumnos)
    promedio_puntaje(puntaje_alumnos, tema_alumnos)
    mayor_puntaje(puntaje_alumnos, tema_alumnos)

# Llamada principal
main()
```

```python
from random import randint
def generar_datos():
    while True:
        try:
            votantes = int(input("ingrese el numero de votantes : "))
            if 1<= votantes <=100:
                break
            else:
                print("el numero de votantes debe ser entre 1 y 100")
        except ValueError:
            print("datos incorrectos")
    lista_votos = []
    lista_dni =[]
    for i in range(votantes):
        votos = randint(1, 3)
        lista_votos.append(votos)
        dni = randint(10000000 , 99999999)
        lista_dni.append(dni)
    return lista_votos, lista_dni

def mostrar_datos(lista_dni , lista_votos):
    print("DNI".center(10) , "VOTO".center(10))
    for i in range(len(lista_dni)):
        print(f"{lista_dni[i]:>10}{lista_votos[i]:>8}")

def candidato_ganador(lista_dni , lista_votos):
    candidato1 = 0
    candidato2 = 0
    for votos in lista_votos:
        if votos == 1:
            candidato1 += 1
        elif votos == 2:
            candidato2 += 1
    print(f"votos candidaro A : {candidato1}")
    print(f"votos candidaro B : {candidato2}")
    if candidato1 == candidato2:
        mayor = "EMPATE ENTRE CANDIDATOS"
    elif candidato1 > candidato2:
        mayor = "CANDIDATO A GANA"
    else:
        mayor = "CANDIDATO B GANA"
    
    print(f"El candidato ganador es :{mayor}")

def dni_votante_longevo(lista_dni):
    for i in range(len(lista_dni)):
        dni_maximo = max(lista_dni)
    return dni_maximo

# programa principal
def main():
    lista_votos , lista_dni = generar_datos()
    mostrar_datos(lista_dni , lista_votos)
    candidato_ganador(lista_dni , lista_votos)
    dni_longevo = dni_votante_longevo(lista_dni)
    print(f"El dni del votante mas viejo es : {dni_longevo}")

main()
        
        
```

```python
from random import choice
from random import randint

def genera_muestra():
    while True:
        try:
            alumnos = int(input("Ingrese el número de estudiantes (máximo 20): "))
            if 1 <= alumnos <= 20:
                break
            else:
                print("Ingrese un número entre 1 y 20")
        except ValueError:
            print("Ingrese un número válido")

    lista_gripe = []
    lista_tos = []
    lista_peso = []
    
    for i in range(alumnos):
        gripe = choice(["S", "N"])
        lista_gripe.append(gripe)
        tos = choice(["S", "N"])
        lista_tos.append(tos)
        peso = randint(30, 60)
        lista_peso.append(peso)
    
    return lista_gripe, lista_tos, alumnos, lista_peso

def mostrar_datos(lista_gripe, lista_tos, alumnos, lista_peso):
    for i in range(alumnos):
        print(f"ALUMNO {i+1}")
        print(f"Gripe: {lista_gripe[i]}, Tos: {lista_tos[i]}, Peso: {lista_peso[i]}")

def porcentaje_gripe_tos(lista_gripe, lista_tos, alumnos):
    porcentaje = 0
    for i in range(alumnos):
        if lista_gripe[i] == "S" and lista_tos[i] == "S":
            porcentaje += 1
    porcent = porcentaje / alumnos * 100
    print(f"El porcentaje de estudiantes con gripe y tos es: {porcent}%")

def promedio_pesos(lista_gripe, lista_tos, lista_peso, alumnos):
    suma = 0
    for i in range(alumnos):
        if lista_gripe[i] == "S" or lista_tos[i] == "S":
            suma += lista_peso[i]
    if suma > 0:
        promedio = suma / alumnos
        print(f"El promedio de peso de los estudiantes con alguna enfermedad es: {round(promedio, 2)}")
    else:
        print("No hay estudiantes con gripe y tos")

def listado_ordenado(lista_gripe, lista_tos, alumnos, lista_peso):
    for i in range(alumnos - 1):
        for j in range(i + 1, alumnos):
            if lista_peso[i] > lista_peso[j]:
                aux = lista_peso[i]
                lista_peso[i] = lista_peso[j]
                lista_peso[j] = aux
    
    # Mostrar datos ordenados
    for i in range(alumnos):
        print(f"ALUMNO {i+1}")
        print(f"Gripe: {lista_gripe[i]}, Tos: {lista_tos[i]}, Peso: {lista_peso[i]}")

def main():
    lista_gripe, lista_tos, alumnos, lista_peso = genera_muestra()
    mostrar_datos(lista_gripe, lista_tos, alumnos, lista_peso)
    porcentaje_gripe_tos(lista_gripe, lista_tos, alumnos)
    promedio_pesos(lista_gripe, lista_tos, lista_peso, alumnos)
    listado_ordenado(lista_gripe, lista_tos, alumnos, lista_peso)

# Ejecutar función
main()
```

```python
from random import randint
from random import choice

def personas_encuestadas():
    while True:
        try:
            personas = int(input("¿Cuántas personas se han encuestado? "))
            if 1 <= personas <= 100:
                break
            else:
                print("Por favor, ingrese un número entre 1 y 100.")
        except ValueError:
            print("Por favor, ingrese un número entero.")
    
    nivel_aprobacion = []
    edad = []
    departamento = []
    for i in range(personas):
        nivel_aprobacion.append(randint(1, 5))
        edad.append(randint(17, 100))
        depart = choice(['A', 'C', 'L', 'P'])
        departamento.append(depart)
    
    return nivel_aprobacion, edad, departamento, personas

def mostrar_datos(nivel_aprobacion, edad, departamento, personas):
    print("DATOS GENERADOS")
    print(f"{'VOTACION'.center(10)} {'EDAD'.center(10)} {'DEPARTAMENTO'.center(15)}")
    for i in range(personas):
        print(f"{nivel_aprobacion[i]:>5} {edad[i]:>10} {departamento[i]:>15}")

def nivel_aprobacion_func(nivel_aprobacion):
    pesimo = 0
    excelente = 0
    for nivel in nivel_aprobacion:
        if nivel == 1:
            pesimo += 1
        elif nivel == 5:
            excelente += 1
            
    print(f"VOTACION POR NIVELES DE APROBACION")
    print(f"Excelente : {excelente}")
    print(f"Pésimo : {pesimo}")

def personas_cajamarca_puno(nivel_aprobacion, edad, departamento):
    cajamarca = 0
    puno = 0
    for dep in departamento:
        if dep == 'C':
            cajamarca += 1
        elif dep == 'P':
            puno += 1
    
    print(f"Personas de Puno: {puno}")
    print(f"Personas de Cajamarca: {cajamarca}")

def ordenar_x_edad(nivel_aprobacion, edad, departamento, personas):
    for i in range(personas - 1):
        for j in range(i + 1, personas):
            if edad[i] < edad[j]:
                aux_edad = edad[i]
                edad[i] = edad[j]
                edad[j] = aux_edad
                aux_dep = departamento[i]
                departamento[i] = departamento[j]
                departamento[j] = aux_dep
                aux_voto = nivel_aprobacion[i]
                nivel_aprobacion[i] = nivel_aprobacion[j]
                nivel_aprobacion[j] = aux_voto
    print("DATOS ORDENADOS POR EDAD")
    print(f"{'VOTACION'.center(10)} {'EDAD'.center(10)} {'DEPARTAMENTO'.center(15)}")
    for i in range(personas):
        print(f"{nivel_aprobacion[i]:>5} {edad[i]:>10} {departamento[i]:>15}")

def mayor_menor_departamento(edad, departamento, personas):
    mayor = max(edad)
    menor = min(edad)
    print(mayor)
    print(f"Departamento con mayor edad: {departamento[edad.index(mayor)]}")
    print(menor)
    print(f"Departamento con menor edad: {departamento[edad.index(menor)]}")

def main():
    nivel_aprobacion, edad, departamento, personas = personas_encuestadas()
    mostrar_datos(nivel_aprobacion, edad, departamento, personas)
    nivel_aprobacion_func(nivel_aprobacion)
    print("\n")
    personas_cajamarca_puno(nivel_aprobacion, edad, departamento)
    print("\n")
    ordenar_x_edad(nivel_aprobacion, edad, departamento, personas)
    print("\n")
    mayor_menor_departamento(edad, departamento, personas)

# Ejecutar función
main()
```

```python
from random import randint 

def registro_datos():
    while True:
        try:
            alumnos = int(input("ingrese el numero de alumnos : "))
            if 1 <= alumnos <= 50:
                break
            else:
                print("cantidad fuera de rango")
        except ValueError:
            print("Datos incorrectos")
    lista_codigo = []
    lista_creditos = []
    promedio_acumulado = []
    for i in range(alumnos):
        lista_codigo.append(randint(10000, 99999))
        lista_creditos.append(randint(1, 50))
        promedio_acumulado.append(randint(1, 20))
    return lista_codigo, lista_creditos, promedio_acumulado, alumnos

def ordenar_x_creditos(lista_codigo, lista_creditos, promedio_acumulado, alumnos):
    for i in range(alumnos - 1):
        for j in range(i + 1, alumnos):
            if lista_creditos[i] > lista_creditos[j]:
                aux = lista_creditos[i]
                lista_creditos[i] = lista_creditos[j]
                lista_creditos[j] = aux
                aux = lista_codigo[i]
                lista_codigo[i] = lista_codigo[j]
                lista_codigo[j] = aux
                aux = promedio_acumulado[i]
                promedio_acumulado[i] = promedio_acumulado[j]
                promedio_acumulado[j] = aux
    print("DATOS ORDENADOS POR CREDITOS".center(20))
    print("______________________________________________________")
    print("codigo".center(10), "Créditos".center(10), "Promedio".center(10))
    for i in range(alumnos):
        print(f"{lista_codigo[i]:>8} {lista_creditos[i]:>8}{promedio_acumulado[i]:>10}")
    print("______________________________________________________")

def ordenar_x_promedio(lista_codigo, lista_creditos, promedio_acumulado, alumnos):
    for i in range(alumnos - 1):
        for j in range(i + 1, alumnos):
            if promedio_acumulado[i] < promedio_acumulado[j]:
                aux = promedio_acumulado[i]
                promedio_acumulado[i] = promedio_acumulado[j]
                promedio_acumulado[j] = aux
                aux = lista_creditos[i]
                lista_creditos[i] = lista_creditos[j]
                lista_creditos[j] = aux
                aux = lista_codigo[i]
                lista_codigo[i] = lista_codigo[j]
                lista_codigo[j] = aux
    print(f"\nDATOS ORDENADOS POR  PROMEDIO".center(20))
    print("______________________________________________________")
    print("codigo".center(10), "Créditos".center(10), "Promedio".center(10))
    for i in range(alumnos):
        print(f"{lista_codigo[i]:>8} {lista_creditos[i]:>8}{promedio_acumulado[i]:>10}")
    print("______________________________________________________")

def alumnos_matriculados(lista_codigo, lista_creditos, promedio_acumulado, alumnos):
    print(f"\nALUMNOS CON MAS DE 15 CREDITOS")
    print("______________________________________________________")
    print("codigo".center(10), "Créditos".center(10), "Promedio".center(10))
    for i in range(alumnos):
        if lista_creditos[i] >= 15:
            print(f"{lista_codigo[i]:>8} {lista_creditos[i]:>8}{promedio_acumulado[i]:>10}")
    print("______________________________________________________")

def alumnos_con_promedio_aprobado(lista_codigo, lista_creditos, promedio_acumulado, alumnos):
    print(f"\nALUMNOS CON PROMEDIO APROBADO")
    print("______________________________________________________")
    print("codigo".center(10), "Créditos".center(10), "Promedio".center(10))
    for i in range(alumnos):
        if promedio_acumulado[i] >= 13:
            print(f"{lista_codigo[i]:>8} {lista_creditos[i]:>8}{promedio_acumulado[i]:>10}")
    print("______________________________________________________")

def main():
    lista_codigo, lista_creditos, promedio_acumulado, alumnos = registro_datos()
    ordenar_x_creditos(lista_codigo, lista_creditos, promedio_acumulado, alumnos)
    ordenar_x_promedio(lista_codigo, lista_creditos, promedio_acumulado, alumnos)
    alumnos_matriculados(lista_codigo, lista_creditos, promedio_acumulado, alumnos)
    alumnos_con_promedio_aprobado(lista_codigo, lista_creditos, promedio_acumulado, alumnos)

# ejecutar programa
main()
```
